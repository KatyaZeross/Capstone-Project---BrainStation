```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 

```

```python
gamezone = pd.read_csv(r"C:\Users\Katya\Documents\data\vgsales.csv")
```

```python
gamezone.head()
```

```python
#plot the ratings count over the years
plt.figure(figsize = (15,10))
plt.bar(gamezone['Year'], gamezone['Global_Sales'], color='darkblue') # it's actually the number of ratings sum up for each year
plt.xlabel('Year', size=15)
plt.ylabel('Global Sales', size=15)
plt.title('Sales Across the Years', size=20)
plt.savefig('Sales Across the Years')
plt.show()
```
